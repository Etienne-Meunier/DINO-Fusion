```python
def normalize_maxmin_4D(data):
    # Get min, max value aming all elements for each column
    x_min = np.min(data[:,:,2:-2,2:-2], axis=(0,2,3), keepdims=True)
    x_max = np.max(data[:,:,2:-2,2:-2], axis=(0,2,3), keepdims=True)
    # Normalize with those min, max values leveraging broadcasting
    data = 2 * ((data - x_min)/ (x_max - x_min)) - 1
    del x_min, x_max
    print("ok")
    return data

def standardize_4D(data):
    """
    data : batch, depth, x, y 
    """
    # Get min, max value aming all elements for each column
    x_mean = np.mean(data[:,:,2:-2,2:-2], axis=(0,2,3), keepdims=True)
    x_std = np.std(data[:,:,2:-2,2:-2], axis=(0,2,3), keepdims=True)
    # Normalize with those min, max values leveraging broadcasting
    data = (data - x_mean)/ x_std
    stand_infos = {"x_mean":x_mean,"x_std":x_std}
    del x_mean, x_std
    print("ok")
    return data, stand_infos
```

```python
fig, axs = plt.subplots(8, 9, figsize=(20, 20))
for i in range(8):
    for j in range(9):
        idx = i * 9 + j 
        if idx < np.shape(data)[0]:
            if idx<=35 : color = "tab:blue"
            if idx>=35 and idx<70 : color = "tab:green"
            if idx==70 : color = "tab:orange"
            axs[i, j].hist(data[idx],color=color)
            axs[i, j].set_title(f'channel {idx}')
plt.tight_layout()
plt.show()
```
